```python
import requests
from bs4 import BeautifulSoup
```

```python
def book_spider(max_pages):
    list_of_books = []
    list_of_authors = []
    page = 1
    while page <= max_pages:    
        url = "https://www.goodreads.com/list/show/276.Best_Book_Titles?page=" + str(page)
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text)
        f = 0
        for name in soup.findAll("span", {"itemprop" : "name"}):
            title = name.string
            if f==0:
                list_of_books.append(title)
                f = 1
            else:
                list_of_authors.append(title)
                f = 0
        page += 1
    return list_of_books, list_of_authors
```

```python
lob, loa = book_spider(5)
```
